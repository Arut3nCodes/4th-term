### What is VCT Masters 2022 Instanbul

This was the second edition of the Valorant Champions, the crowning event of the  **Valorant Champions Tour**  (VCT) for the 2022 Valorant competitive season. The tournament was held from August 31 to September 18, 2022, in Istanbul, Turkey. 2022 Valorant Champions.

### Plots:
 - Role participation in kills across the whole tournament
 - Most common players nationalities
 - Scatter Death to Kills for every single player
 - K/D of all players for every single team participating in the event
 - Headshot percentage, Kills per Round and K/D for top fraggers for this event

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

data = pd.read_csv('valorant champions istanbul.csv')
data.sample(4)
```

```python
KPPerRole = data.groupby('Role')['Kill'].sum()
KPPerRole.info()
```

```python
KPPerRole.sample(5)
```

```python
plot = KPPerRole.plot.pie(y='', x='', figsize=(11, 6), autopct='%.2f') 

plot.legend(title="Roles",
       loc="center left",
          bbox_to_anchor=(1, 0.05, 0.5, 1))

plot.set_title("KP% for all the roles during VCT Masters Instanbul 2022")
plt.show()
```

```python
howManyPlayersFromNation = data['Nationality'].value_counts()
display(howManyPlayersFromNation)
```

```python
display(howManyPlayersFromNation.head(5))
```

```python
howManyPlayersFromNation.info()
```

```python
display(howManyPlayersFromNation)
print(howManyPlayersFromNation.keys())
```

```python
plot = howManyPlayersFromNation.head(5).plot(kind="bar")
plt.title("Most common players' nationalities")
plot.set_xlabel("Nationality")
plot.set_ylabel("Number of players")
plt.show()
```

```python
KillsToDeaths = data.sort_values(by="Rounds Played")
display(KillsToDeaths)
```

```python
subplot = plt.subplot()
subplot.scatter(x=KillsToDeaths['Death'], y=KillsToDeaths['Kill'])
subplot.set_xlabel('Deaths')
subplot.set_ylabel('Kills')
plt.title("Kills to deaths of every single player")
plt.show()
```

```python
listOfUniques = data['Team'].unique()
figure, axis = plt.subplots(4, 2)
x=0
y=0
figure.suptitle('K/D for all the teams attending VCT Masters 2022', fontsize=30)
figure.set_figheight(23)
figure.set_figwidth(20)

colors = plt.rcParams["axes.prop_cycle"]()

for unique in listOfUniques:
    tempdf = data[ data['Team'] == unique]
    c = next(colors)["color"]
    axis[x][y].set_title(f'{unique}', fontsize=20)
    axis[x][y].bar(tempdf['Player'], tempdf['K/D'], label=unique, color=c)
    y+=1
    if(y==2):
        x+=1
        y=0
plt.show()
```

```python
#Existing data frame with Kills per Rounds to measure efficiency of a player
dataWithKPR = data.__deepcopy__()
dataWithKPR['KPR'] = dataWithKPR['Kill'] / dataWithKPR['Rounds Played']
dataWithKPR['Player'] = dataWithKPR['Team'] + " " + dataWithKPR['Player']
dataWithKPR = dataWithKPR.set_index('Player')
display(dataWithKPR)
```

```python
dataWithKPR = dataWithKPR.sort_values(by='KPR', ascending=False)
display(dataWithKPR.head(5))
```

```python
#Derke bald buff
figure, axis = plt.subplots(1, 2)
figure.suptitle('K/D, HS% and KPR for top 10 VCT Masters fraggers', fontsize=20)
figure.set_figheight(5)
figure.set_figwidth(10)

dataWithKPR = dataWithKPR.sort_values(by='KPR', ascending=False)
dataWithKPR[['KPR', 'K/D']].head(10).plot(kind='bar', ax=axis[0], title='K/D and KPR')
dataWithKPR['HS %'].head(10).plot(kind='bar', ax=axis[1], title='HS%')
plt.show()
```
