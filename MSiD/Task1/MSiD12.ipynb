### What is VCT Masters 2022 Instanbul

This was the second edition of the Valorant Champions, the crowning event of the  **Valorant Champions Tour**  (VCT) for the 2022 Valorant competitive season. The tournament was held from August 31 to September 18, 2022, in Istanbul, Turkey. 2022 Valorant Champions.

### Plots:
 - Role participation in kills across the whole tournament
 - Most common players nationalities
 - Scatter Death to Kills for every single player
 - K/D of all players for every single team participating in the event
 - Headshot percentage, Kills per Round and K/D for top fraggers for this event

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

data = pd.read_csv('valorant champions istanbul.csv')
data.sample(4)
```

```python
KPPerRole = data.groupby('Role')['Kill'].sum()
KPPerRole.info()
```

```python
KPPerRole.sample(5)
```

```python
plot = KPPerRole.plot.pie(y='', x='', figsize=(11, 6), autopct='%.2f') 

plot.legend(title="Roles",
       loc="center left",
          bbox_to_anchor=(1, 0.05, 0.5, 1))

plot.set_title("KP% for all the roles during VCT Masters Instanbul 2022")
plt.show()
```

```python
howManyPlayersFromNation = data['Nationality'].value_counts()
display(howManyPlayersFromNation)
```

```python
display(howManyPlayersFromNation.head(5))
```

```python
howManyPlayersFromNation.info()
```

```python
display(howManyPlayersFromNation)
print(howManyPlayersFromNation.keys())
```

```python
plot = howManyPlayersFromNation.head(5).plot(kind="bar")
plt.title("Most common players' nationalities")
plot.set_xlabel("Nationality")
plot.set_ylabel("Number of players")
plt.show()
```

```python
KillsToDeaths = data.sort_values(by="Rounds Played")
display(KillsToDeaths)
```

```python
subplot = plt.subplot()
subplot.scatter(x=KillsToDeaths['Death'], y=KillsToDeaths['Kill'])
subplot.set_xlabel('Deaths')
subplot.set_ylabel('Kills')
plt.title("Kills to deaths of every single player")
plt.show()
```

```python
listOfUniques = data['Team'].unique()
figure, axis = plt.subplots(4, 2)
x=0
y=0
figure.suptitle('K/D for all the teams attending VCT Masters 2022', fontsize=30)
figure.set_figheight(23)
figure.set_figwidth(20)

colors = plt.rcParams["axes.prop_cycle"]()

for unique in listOfUniques:
    tempdf = data[ data['Team'] == unique]
    c = next(colors)["color"]
    axis[x][y].set_title(f'{unique}', fontsize=20)
    axis[x][y].bar(tempdf['Player'], tempdf['K/D'], label=unique, color=c)
    y+=1
    if(y==2):
        x+=1
        y=0
plt.show()
```

```python
#Existing data frame with Kills per Rounds to measure efficiency of a player
dataWithKPR = data.__deepcopy__()
dataWithKPR['KPR'] = dataWithKPR['Kill'] / dataWithKPR['Rounds Played']
dataWithKPR['Player'] = dataWithKPR['Team'] + " " + dataWithKPR['Player']
dataWithKPR = dataWithKPR.set_index('Player')
display(dataWithKPR)
```

```python
dataWithKPR = dataWithKPR.sort_values(by='KPR', ascending=False)
display(dataWithKPR.head(5))
```

```python
#Derke bald buff
figure, axis = plt.subplots(1, 2)
figure.suptitle('K/D, HS% and KPR for top 10 VCT Masters fraggers', fontsize=20)
figure.set_figheight(5)
figure.set_figwidth(10)

dataWithKPR = dataWithKPR.sort_values(by='KPR', ascending=False)
dataWithKPR[['KPR', 'K/D']].head(10).plot(kind='bar', ax=axis[0], title='K/D and KPR')
dataWithKPR['HS %'].head(10).plot(kind='bar', ax=axis[1], title='HS%')
plt.show()
```

### Do dalszego obrabiania danych wykorzystam wykres z poprzedniego zadania:

```python
KillsToDeaths = data.sort_values(by="Rounds Played")
subplot = plt.subplot()
subplot.scatter(x=KillsToDeaths['Death'], y=KillsToDeaths['Kill'])
subplot.set_xlabel('Deaths')
subplot.set_ylabel('Kills')
plt.title("Kills to deaths of every single player")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
modelLin = LinearRegression()
xAdj = KillsToDeaths['Death'].values
yAdj = KillsToDeaths['Kill'].values
print(xAdj, yAdj)
```

```python
modelLin.fit(xAdj.reshape(-1,1), yAdj)
```

```python
X_test = np.linspace(start=xAdj.min(), stop=xAdj.max(), num=40)
Y_pred = modelLin.predict(X_test.reshape(-1,1))

plt.scatter(xAdj, yAdj, alpha=0.7)
plt.plot(X_test, Y_pred, color='tab:orange', linewidth=2)
plt.title("Kills to deaths of every single player")
plt.xlabel('Deaths')
plt.ylabel('Kills')
plt.show()
```

```python
print(f'Parametry modelu liniowego: a = {np.round(modelLin.coef_,5)}, b = {np.round(modelLin.intercept_,5)}')
```

#### Przeprowadźmy test dla samych wartości:

```python
X = xAdj
Y = yAdj

# =========  Model liniowy =========
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
modelLin = LinearRegression()
modelLin.fit(X.reshape(-1,1), 
              Y)
print(f'Parametry modelu liniowego: {np.round(modelLin.coef_,5)}, {np.round(modelLin.intercept_,5)}')
MSE_lin = mean_squared_error(Y, modelLin.predict(X.reshape(-1,1)))
print(f'Błąd średniokwadratowy modelu liniowego: {MSE_lin:0.3}\n')

# ===== Uogólniony model liniowy (ang. Generalized Linear Model) =====
from sklearn.preprocessing import PolynomialFeatures
model_GLM = LinearRegression()
gen_features = PolynomialFeatures(degree=2, include_bias=True, interaction_only=False)
model_GLM.fit(gen_features.fit_transform(X.reshape(-1,1)), 
              Y)
print(f'Parametry modelu GLM: {np.round(model_GLM.coef_,4)}, {np.round(model_GLM.intercept_,5)}')
MSE_GLM = mean_squared_error(Y, model_GLM.predict(gen_features.fit_transform(X.reshape(-1,1))))
print(f'Błąd średniokwadratowy modelu GLM: {MSE_GLM:0.3}\n')

#==== Maszyna wektorów wspierających (ang. Support Vector Machine) ====
# SVR dla regresji, SVC dla kalsyfikacji
from sklearn.svm import SVR
model_svr = SVR(kernel='rbf', gamma='scale', C=1)
model_svr.fit(X.reshape(-1,1), 
              Y)
MSE_SVR = mean_squared_error(Y, model_svr.predict(X.reshape(-1,1)))
print(f'Błąd średniokwadratowy modelu SVR: {MSE_SVR:0.3}')

# Predykcje wszystkich modeli dla całego zakreso osi X
os_x = np.linspace(start=X.min(), stop=X.max(), num=300)
y_lin_pred = modelLin.predict(os_x.reshape(-1,1))
y_GLM_pred = model_GLM.predict(gen_features.fit_transform(os_x.reshape(-1,1)))
y_svr_pred = model_svr.predict(os_x.reshape(-1,1))

# Wizualizacja
plt.figure(figsize=(10,7))
plt.scatter(X, Y, label='Dane', alpha=0.7)
plt.plot(os_x, y_lin_pred, label='model liniowy', color='tab:orange')
plt.plot(os_x, y_GLM_pred, label=f'model GLM', color='tab:red')
plt.plot(os_x, y_svr_pred, label='model SVR', color='tab:green')
plt.title("Kills to deaths of every single player")
plt.xlabel('Deaths')
plt.ylabel('Kills')
plt.legend(fontsize=12, shadow=True, loc='lower right')
plt.ylim([Y.min()-0.1, Y.max()+0.5])
plt.show()
```

```python
X = xAdj
Y = yAdj

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y)

# =========  Model liniowy =========
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
modelLin = LinearRegression()
modelLin.fit(X_train.reshape(-1,1), 
              Y_train)
print(f'Parametry modelu liniowego: {np.round(modelLin.coef_,5)}, {np.round(modelLin.intercept_,5)}')
MSE_lin = mean_squared_error(Y_test, modelLin.predict(X_test.reshape(-1,1)))
print(f'Błąd średniokwadratowy modelu liniowego: {MSE_lin:0.3}\n')

# ===== Uogólniony model liniowy (ang. Generalized Linear Model) =====
from sklearn.preprocessing import PolynomialFeatures
model_GLM = LinearRegression()
gen_features = PolynomialFeatures(degree=2, include_bias=True, interaction_only=False)
model_GLM.fit(gen_features.fit_transform(X_train.reshape(-1,1)), 
              Y_train)
print(f'Parametry modelu GLM: {np.round(model_GLM.coef_,4)}, {np.round(model_GLM.intercept_,5)}')
MSE_GLM = mean_squared_error(Y_test, model_GLM.predict(gen_features.fit_transform(X_test.reshape(-1,1))))
print(f'Błąd średniokwadratowy modelu GLM: {MSE_GLM:0.3}\n')

#==== Maszyna wektorów wspierających (ang. Support Vector Machine) ====
# SVR dla regresji, SVC dla kalsyfikacji
from sklearn.svm import SVR
model_svr = SVR(kernel='rbf', gamma='scale', C=1)
model_svr.fit(X_train.reshape(-1,1), 
              Y_train)
MSE_SVR = mean_squared_error(Y_test, model_svr.predict(X_test.reshape(-1,1)))
print(f'Błąd średniokwadratowy modelu SVR: {MSE_SVR:0.3}')

# Predykcje wszystkich modeli dla całego zakreso osi X
os_x = np.linspace(start=X.min(), stop=X.max(), num=300)
y_lin_pred = modelLin.predict(os_x.reshape(-1,1))
y_GLM_pred = model_GLM.predict(gen_features.fit_transform(os_x.reshape(-1,1)))
y_svr_pred = model_svr.predict(os_x.reshape(-1,1))

# Wizualizacja
plt.figure(figsize=(10,7))
plt.scatter(X_train, Y_train, label='dane treningowe', alpha=0.7)
plt.scatter(X_test, Y_test, edgecolor='black', facecolor='none', label='dane testujące')
plt.plot(os_x, y_lin_pred, label='model liniowy', color='tab:orange')
plt.plot(os_x, y_GLM_pred, label=f'model GLM', color='tab:red')
plt.plot(os_x, y_svr_pred, label='model SVR', color='tab:green')
plt.title("Kills to deaths of every single player")
plt.xlabel('Deaths')
plt.ylabel('Kills')
plt.legend(fontsize=12, shadow=True, loc='lower right')
plt.ylim([Y.min()-0.1, Y.max()+0.5])
plt.show()
```
